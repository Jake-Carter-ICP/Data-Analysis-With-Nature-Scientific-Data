```python
def filt_to_hist(df_to_plot, crop_filter_var, hist_var):
    """Function filters to filter var then plots hist_var as
    a histogram"""
    df_in = df_to_plot \
        .query('Crop == @crop_filter_var')
    
    fig, ax = plt.subplots(figsize = [6,4], layout = 'constrained')

    ax.hist(plot_data[hist_var], bins = 40, edgecolor = 'black')
    ax.set_title('Crop = ' + crop_filter_var + '\n Var = ' + hist_var)
    plt.show()
```
