```python
def filt_to_hist(df_to_plot, crop_filter_var, hist_var):
    """Function filters to filter var then plots hist_var as
    a histogram"""
    df_in = df_to_plot \
        .query('Crop == @crop_filter_var')
    
    fig, ax = plt.subplots(figsize = [6,4], layout = 'constrained')

    ax.hist(plot_data[hist_var], bins = 40, edgecolor = 'black')
    ax.set_title('Crop = ' + crop_filter_var + '\n Var = ' + hist_var)
    plt.show()
```

```python
filt_to_hist(df, 'Wheat', 'Ca_grain')
```

Python's summary functions automatically ignores NAs

```python
df_summary =\
df_long\
    .groupby('object', as_index = False)\
    .agg({'value': ['mean', 'median', 'std', 'size', 'count', lambda x: x.isna().sum()]})

df_summary.rename(columns={'<lambda_0>': 'n_NA'}, inplace=True)

df_summary.columns.to_flat_index()
df_summary.columns = ["_".join(c) for c in df_summary.columns]
df_summary.rename(columns = {'value_count':'value_n_no_NA'}, inplace=True)
df_summary
```
